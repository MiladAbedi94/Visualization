```r
suppressMessages(library(feather))
suppressMessages(library(tidyverse))
suppressMessages(library(ggplot2))
suppressMessages(library(ggExtra))
suppressMessages(library(repr))
```

```r
df_dataset<-read_feather("DataSet/DataSet.feather")
df_reviews<-read_feather("DataSet/Reviews.feather")
```

```r
options(repr.plot.width = 6, repr.plot.height =5)
labeller_list=c("TRUE"="Host is a superhost","FALSE"="Host is not a superhost",'Private room'='Private room','Entire home/apt'='Entire home/apt','Shared room'='Shared room')
df_dataset %>% 
    subset((!is.na(host_is_superhost)) & price<2500 & !is.na(review_scores_rating) ) %>%
        ggplot(mapping=aes(x=price/100,y=.$review_scores_rating,color=host_is_superhost,stroke=0))+
        geom_jitter(alpha=0.5)+
        facet_wrap(host_is_superhost~room_type,labeller = as_labeller(labeller_list))+
        xlab("Price (100$)")+
        ylab("Review Rating")+
        theme(element_line(size = 0.15))+
        labs(color="Superhost Status")+
        guides(color=FALSE)
```

```r
df_dataset%>%
    subset((!is.na(host_is_superhost)) & price<2500)%>%
        ggplot(mapping=aes(x=price,fill=host_is_superhost,color=host_is_superhost))+
        geom_histogram(mapping=aes(y=..density..),alpha=0.5,position="identity",lwd=.4,bins=40)+
        xlab("Price")+
        ylab("Density")+
        guides(fill = guide_legend(title = "Superhost?"),color=FALSE)
```

```r
options(repr.plot.width = 4, repr.plot.height =4)
p=ggplot(data=df_dataset,mapping=aes(x=review_scores_rating,y=df_dataset$"avg(Sentiment_Score)",stroke=0))+geom_jitter(color="#00BCD8")+xlab("Average Rating")+ylab("Average Sentiment Score")
ggMarginal(p,type="histogram",colour='#B983FF',fill='#B983FF',bins=50)


```
